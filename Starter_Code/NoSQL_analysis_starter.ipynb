```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
# Use count_documents to display the number of documents in the result
establishments_count = establishments.count_documents(filter=query)
print(establishments_count)
# Display the first document in the results using pprint
pprint(establishments.find_one(filter=query))
```

```python
# Convert the result to a Pandas DataFrame
results = list(establishments.find(query))
result_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(len(result_df))
# Display the first 10 rows of the DataFrame
result_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"AddressLine2": "London",
         "RatingValue": {'$gte': 4}}

# Use count_documents to display the number of documents in the result
establishments_count = establishments.count_documents(filter=query)
print(establishments_count)
# Display the first document in the results using pprint
for result in establishments.find(filter=query):
    pprint(result)
    break
```

```python
# Convert the result to a Pandas DataFrame
results = list(establishments.find(query))
results_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(len(results_df))
# Display the first 10 rows of the DataFrame
results_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
CoordinatesQuery = {"AddressLine1": "Penang Flavours"}

CoordinatesResults = establishments.find(CoordinatesQuery)
    
degree_search = 0.01
latitude = CoordinatesResults[0]["geocode"]["latitude"].to_decimal()
longitude = CoordinatesResults[0]["geocode"]["longitude"].to_decimal()

latitudeMin = float(latitude) - degree_search
latitudeMax = float(latitude) + degree_search
longitudeMin = float(longitude) - degree_search
longitudeMax = float(longitude) + degree_search


query = {"geocode.latitude": {"$gte": latitudeMin,
                              "$lte": latitudeMax},
         "geocode.longitude": {"$gte": longitudeMin,
                              "$lte": longitudeMax},
         "RatingValue": {"$eq": 5}}

fields = {"BusinessName": 1,
          "RatingValue": 1,
          "geocode.latitude": 1,
          "geocode.longitude": 1,
          "scores.Hygiene": 1}

sort = [('scores.Hygiene', 1)]
limit = 5

# Print the results
results = list(establishments.find(query, fields).sort(sort).limit(limit))

pprint(results)
```

```python
# Convert result to Pandas DataFrame
results = list(establishments.find(query, fields).sort(sort).limit(limit))
rating_value_5_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rating_value_5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}
# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1}}
pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))
# Print the number of documents in the result
print(len(results))
# Print the first 10 results
i = 0
for result in results:
    pprint(result)
    i += 1
    if i == 10:
        break
    
# Cast the results as a list and save the results to a variable
establishments_list = [result for result in results]
# Pretty print the results
pprint(establishments_list)
```

```python
# Convert the result to a Pandas DataFrame
zero_hygiene_df = pd.DataFrame(establishments_list)
# Display the number of rows in the DataFrame
print(len(zero_hygiene_df))
# Display the first 10 rows of the DataFrame
zero_hygiene_df.head(10)
```
